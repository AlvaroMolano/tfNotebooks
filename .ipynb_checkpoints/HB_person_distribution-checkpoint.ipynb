```python
# ETCBC's BHSA data
from tf.fabric import Fabric
from tf.extra.bhsa import Bhsa

# stats & data-containers
import collections
import pandas as pd
import numpy as np
import scipy.stats as stats

# data visualizations
import seaborn as sns
sns.set(style="whitegrid")
import matplotlib.pyplot as plt

# prep the data
name = 'HB_person_distribution'
hebrew_data = ['~/github/etcbc/{}/tf/c'.format(direc) for direc in ('bhsa',)] # data dirs
load_features = '''
ps prs_ps domain
''' 

# TF load statements
TF = Fabric(locations=hebrew_data, silent=True)
api = TF.load(load_features)
B = Bhsa(api=api, hoist=globals(), silent=True, name=name) # Bhsa functions for search and visualizing text
```

```python
person_results = list()

header = ('node', 'person', 'book', 'domain')

for word in F.otype.s('word'):
    
    domain = F.domain.v(L.u(word, 'clause')[0])
    person = F.ps.v(word)
    person_suffix = F.prs_ps.v(word)
    persons = [p for p in (person, person_suffix) if all([p not in {'unknown', 'NA'},
                                                          domain in {'Q', 'N'}
                                                         ])]
    
    for person in persons:
        book, chapter, verse = T.sectionFromNode(word)
        person_results.append([word, person, book, domain])
        
person_results = pd.DataFrame(person_results, columns=header)
        
print(f'Done with {len(person_results)} results...')
print('Sample:')
person_results.head()
```

```python
def show_data():
    plt.figure(figsize=(10, 6))
    sns.set(font_scale=1.6)
    sns.countplot(x='domain', hue='person', data=person_results).set_title('Person Counts Across the Hebrew Bible by Narrative/Quotation')
```
