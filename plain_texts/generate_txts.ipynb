## Pentateuch

```python
bhsa = use('bhsa:clone').api
```

```python
pent = '''

### Excerpt from BHSA Repository
Dirk Roorda and the Eep Talstra Centre for Bible and Computer.
https://github.com/ETCBC/bhsa

This is the text-fabric representation of the Hebrew Bible Database, containing the text of the Hebrew Bible augmented with linguistic annotations compiled by the Eep Talstra Centre for Bible and Computer, VU University Amsterdam.

This repository contains several versions of the dataset, going back to 2011. It is meant to accumulate new versions over the years, and have one continuous version, a version that we deliberately update with a high frequency.

So this repository addresses the need for permanency as well as the urge for change.

More information, especially about the data versions and annotation features, can be found on the data documentation pages.

See also

Coding the Hebrew Bible
License

This work is licensed under a Attribution-NonCommercial 4.0 International (CC BY-NC 4.0). That means:

You may download the data and use it: process, copy, modify;
You may use the data to create new software applications;
You may use the data for research and publish any amount of results;
When you publish this data or results you obtained from them, you have to comply with the following:
give proper attribution to the data when you use it in new applications, by citing this persistent identifier: 10.17026/dans-z6y-skyh.
do not use the data for commercial applications without consent; for any commercial use, please contact the German Bible Society.

'''

line = ''

for book in bhsa.F.otype.s('book'):
    
    book_name = bhsa.Fs('book@en').v(book)
    
    if book_name not in {'Genesis', 'Exodus', 'Leviticus', 'Numbers','Deuteronomy'}:
        continue
    
    pent += f'\n\n{book_name}'
    
    for chapter in bhsa.L.d(book,'chapter'):
        pent += f'\n\n{bhsa.F.chapter.v(chapter)}\n'

        for verse in bhsa.L.d(book, 'verse'):
            #line += f'{bhsa.F.verse.v(verse)} '
        
            for word in bhsa.L.d(verse,'word'):
                if len(line) > 160:
                    pent += line+'\n'
                    line = bhsa.T.text(word)
                else:
                    line += bhsa.T.text(word)
                    
        pent += line + '\n'
        line = ''
    
pent += '''

** This is the end of the Pentateuch ** 

'''

with open(texts_loca+'hebrew_Pentateuch.txt', 'w') as outfile:
    outfile.write(pent)
```

## GNT

```python
gnt = use('tisch:local').api
```

```python
gnt_text = '''
            Tischendorf's 8th edition Greek New Testament
                       with morphological tags
                             Version X.X

              Based on G. Clint Yale's Tischendorf text
                   and on Dr. Maurice A. Robinson's
                   Public Domain Westcott-Hort text

                  Edited by Ulrik Sandborg-Petersen

                    This text and its analysis are 
                  in the Public Domain. Copy freely.


-- Text-Fabric Source Notes -- 
https://github.com/codykingham/tischendorf_tf

This text corpus is made available by the work of Dr. Ulrik Sandborg-Petersen of Scripture Systems Aps, Denmark with help from G. Clint Yale and Dr. Maurice A. Robinson.

The conversion to TF is carried out by Cody Kingham and Oliver Glanz.

The Text-Fabric corpus tool was designed by Dirk Roorda of DANS.
'''

line = ''

for book in gnt.F.otype.s('book'):
    
    book_name = gnt.F.book.v(book)
    
    if book_name not in {'Matthew', 'Mark', 'Luke', 'John'}:
        continue
    
    gnt_text += f'\n\n{book_name}'
    
    for chapter in gnt.L.d(book,'chapter'):
        gnt_text += f'\n\n{gnt.F.chapter.v(chapter)}\n'

        for verse in gnt.L.d(chapter, 'verse'):
            line += f'({gnt.F.verse.v(verse)}) '
        
            for word in gnt.L.d(verse,'word'):
                if len(line) > 80:
                    gnt_text += line+'\n'
                    line = gnt.T.text(word)
                else:
                    line += gnt.T.text(word)
                    
        gnt_text += line + '\n'
        line = ''
        
gnt_text += '\n\n**This is the end of the GNT **'


with open(texts_loca+'greek_Gospels.txt', 'w') as outfile:
    outfile.write(gnt_text)
```

## NENA

```python
TF = Fabric('~/github/CambridgeSemiticsLab/nena_tf/tf/0.02')
nena = TF.load('title number')
```

```python
barwar = nena.T.nodeFromSection(('Barwar',))
```

```python
nena_text = """

Northeastern Neo-Aramaic consists of a very diverse group of Aramaic dialects that were spoken until modern times in Northern Iraq, North West Iran and South Eastern Turkey by Christian and Jewish communities. These are among the last remaining living vestiges of the Aramaic language, which was one of the major languages of the region in antiquity.

This text corpus consists of transcribed and recorded texts gathered by Prof. Geoffrey Khan and his team in their efforts to preserve these increasingly endangered languages. This repository contains raw and corrected source texts written in North Eastern Neo-Aramaic (NENA) dialects. All original source material is converted to a standard mark-up format, regardless of its original format.

The purpose of this repository is to curate source texts that will be used for building a complete text corpus in Text-Fabric. The corpus will in turn be analyzed and annotated for linguistic features.

LICENSE: 
Creative Commons International 4.0

Source: 
https://github.com/CambridgeSemiticsLab/nena_corpus

This is the Barwar corpus.
"""

block = ''

for text in nena.L.d(barwar, 'text'):
    
    text_name = nena.F.title.v(text)
    nena_text += f'\n\n{text_name}\n\n'
    
    for line in nena.L.d(text,'line'):
        block += f'({nena.F.number.v(line)}) '
    
        for word in nena.L.d(line, 'word'):

            if len(block) > 80:
                nena_text += block+'\n'
                block = nena.T.text(word, fmt='text-orig-lite')
            else:
                block += nena.T.text(word, fmt='text-orig-lite')
                
    nena_text += block + '\n'
    block = ''
    
nena_text += '''

-------------

This is the end of the Barwar corpus.

'''
with open(texts_loca+'NENA_barwar.txt', 'w') as outfile:
    outfile.write(nena_text)
```
